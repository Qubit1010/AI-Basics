```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Building the CNN Architecture
def create_cnn_model(input_shape=(28, 28, 1), num_classes=10):
    model = models.Sequential([
        # First Convolutional Layer: 32 filters, 3x3 kernel size
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        # Pooling Layer: Reduces spatial dimensions by half
        layers.MaxPooling2D((2, 2)),

        # Second Convolutional Layer: Increases complexity
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        # Flattening: Converting 2D maps into a 1D vector
        layers.Flatten(),

        # Fully Connected (Dense) Layers for classification
        layers.Dense(64, activation='relu'),
        layers.Dense(num_classes, activation='softmax') # Softmax for probability distribution
    ])

    return model

# 2. Compiling the model
model = create_cnn_model()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Displaying the summary of the network
model.summary()
```
